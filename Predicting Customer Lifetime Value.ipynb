The mean squared error (MSE) is a measure of the average squared difference between the predicted values and the actual values. A smaller MSE indicates a better fit of the model to the data.

The R-squared (R2) is a statistical measure that represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model. It measures how well the regression line approximates the real data points.
